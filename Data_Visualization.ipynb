```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
df1 = pd.read_csv("Dataset .csv")
```

```python
df1.info()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df1["Aggregate rating"], bins=20, kde=True, color="skyblue")
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x="Aggregate rating", data=df1, palette="cividis")
plt.title("Count of Ratings")
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="Aggregate rating", data=df1, color="skyblue")
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
average_rating_by_city = (
    df1.groupby("City")["Aggregate rating"]
    .mean()
    .sort_values(ascending=False)
)

plt.figure(figsize=(12, 6))
sns.barplot(
    x=average_rating_by_city.head(10).index,
    y=average_rating_by_city.head(10).values,
    palette="viridis"
)
plt.title("Average Ratings of Different Cities (Top 10)")
plt.xlabel("City")
plt.ylabel("Average Ratings")
plt.xticks(rotation=45)
plt.show()
```

```python
features = [
    "Average Cost for two",
    "Votes",
    "Price range",
    "Has Table booking",
    "Has Online delivery",
    "Aggregate rating"
]

sns.pairplot(df1[features])
plt.show()
```
