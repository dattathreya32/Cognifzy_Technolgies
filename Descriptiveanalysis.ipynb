```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(color_codes=True)
%matplotlib inline
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df[["Average Cost for two", "Price range", "Aggregate rating", "Votes"]].describe()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="Country Code", data=df, palette="cividis")

plt.title("Distribution of Restarants by Country Code")
plt.xlabel("country code")
plt.ylabel("number of Resturants")
plt.show()
```

```python
top_countries = df["Country Code"].value_counts().head()

print("Top 5 Countries with the Highest Number of Restarants :")
print(top_countries)
```

```python
plt.figure(figsize=(15,6))

sns.countplot(x="City",data=df, order=df["City"].value_counts().head(20).index, palette="Set2")

plt.title("Distrubution of Restarants by City :")
plt.xlabel("City")
plt.ylabel("Number of Restarants")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,6))

cuisines_count = df['Cuisines'].value_counts()
cuisines_count.head(20).plot(kind="bar", color=sns.color_palette("Set2"))

plt.title("Top 2- Cuisines with the Highest Number of Restarants :")
plt.xlabel("Cuisine")
plt.ylabel("Number of Resturants")
plt.xticks(rotation=45)
plt.show()
```

```python
top_cities = df["City"].value_counts().head(10)

print("Top 10 Cities with the Highest Number of Restarants :")
print(top_cities)
```

```python
top_cuisines = cuisines_count.head(10)
print("Top 10 Cuisines with the Highest Number of Restarants :")
print(top_cuisines)
```
